```python
import pandas as pd
df=pd.read_csv("/myntra_dataset_ByScraping.csv")
df.duplicated().sum()
df=df.drop_duplicates()
df.shape
```

```python
#bar graph
```

```python
import matplotlib.pyplot as plt
top_brands=df["brand_name"].value_counts().head(10)
plt.figure(figsize=(10,5))
top_brands.plot(kind="bar")
plt.title("Top 10 brands by Number of Products")
plt.xlabel("Brand Name")
plt.ylabel("Number of Products")

plt.show()
```

```python
import matplotlib.pyplot as plt
#highest brand discount
brand_discount = df.groupby("brand_name")["discount_percent"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

brand_discount.plot(kind="bar")

plt.title("Top 10 Brands with Highest Average Discount")
plt.xlabel("Brand")
plt.ylabel("Average Discount")

plt.show()
plt.close()
```

#for brand ratings

```python
brand_ratings=df.groupby("brand_name")["ratings"].mean().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
brand_ratings=df.groupby("brand_name")["ratings"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
brand_ratings.plot(kind="bar")
plt.title("Top 10 Brands by Average Ratings")
plt.xlabel("Brand")
plt.ylabel("Average Ratings")
plt.show()
plt.close()
```

#price

```python
brand_price=df.groupby("brand_name")["price"].mean().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
brand_price=df.groupby("brand_name")["price"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
brand_price.plot(kind="bar",color="green")
plt.title("Top 10 Brands by Average Price")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.show()
```

```python
#correlation corr #colormatrix or colorbar
```

```python
df[["price","MRP","discount_percent","ratings","number_of_ratings"]].corr()
```

```python
import matplotlib.pyplot as plt
correlation=df[["price","MRP","discount_percent","ratings","number_of_ratings"]].corr()
plt.figure(figsize=(9,7))
plt.imshow(correlation,cmap="Blues")
plt.colorbar()
plt.title("Correlation Matrix")
plt.xticks(range(len(correlation.columns)),correlation.columns)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.show()

```

```python
#scatter plot
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.scatter(df["price"], df["ratings"], alpha=0.3,color="green")
plt.title("Price vs Ratings")
plt.xlabel("Price")
plt.ylabel("Ratings")
plt.show()
```
